#### Merging tables

```python
# dropping 'Unnamed: 0' column from all tables
reviews_0_250 = reviews_0_250.drop(columns=['Unnamed: 0'])
reviews_250_500 = reviews_250_500.drop(columns=['Unnamed: 0'])
reviews_500_750 = reviews_500_750.drop(columns=['Unnamed: 0'])
reviews_1250_end = reviews_1250_end.drop(columns=['Unnamed: 0'])
```

**Merging (stacking) all review tables together**

```python
reviews_all = pd.concat([
    reviews_0_250,
    reviews_250_500,
    reviews_500_750,
    reviews_1250_end
], ignore_index=True)

print(reviews_all.shape)
reviews_all.head()
```

```python
# downloading merged reviews to csv file
# reviews_all.to_csv("reviews_all.csv", index=False)
```

**Merging reviews with product tables**

```python
full_data = pd.merge(
    reviews_all,
    products,
    on='product_id',
    how='left'
)

print(full_data.shape)
full_data.head()
```

Note: Since we merged the full review tables and products, both tables had rating, product_name, brand_name, and price_usd. So after the merging, pandas automatically assigned:
- From reviews table: rating_x, product_name_x, brand_name_x, price_usd_x
- From products table: rating_y, product_name_y, brand_name_y, price_usd_y

```python
full_data.info()
full_data.isnull().sum()
full_data['rating_x'].value_counts()
```
